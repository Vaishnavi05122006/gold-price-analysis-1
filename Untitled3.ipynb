```python
# GOLD PRICE TREND ANALYSIS - Mobile Friendly Version
# Paste this ENTIRE block into ONE cell in Colab, then run it.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://github.com/datasets/gold-prices/raw/refs/heads/main/data/monthly.csv"
df = pd.read_csv(url)
print(df.head())
print(df.tail())

print("Dataset shape:", df.shape)
print(df.isnull().sum())

df['Date'] = pd.to_datetime(df['Date'])
df = df.dropna(subset=['Price'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Price'])
plt.title('Gold Price Trend (Full History)')
plt.xlabel('Year')
plt.ylabel('Price (USD per troy ounce)')
plt.show()

recent = df[df['Year'] >= 2000]
plt.figure(figsize=(10,5))
plt.plot(recent['Date'], recent['Price'], color='goldenrod')
plt.title('Gold Price Trend (2000 - Present)')
plt.xlabel('Year')
plt.ylabel('Price (USD per troy ounce)')
plt.show()

yearly_avg = df.groupby('Year')['Price'].mean()
print("Yearly average price (last 15 years):")
print(yearly_avg.tail(15))

df['YoY_change'] = df.groupby('Month')['Price'].pct_change() * 100
last_12 = df.tail(12)
plt.figure(figsize=(8,5))
plt.bar(last_12['Date'].dt.strftime('%Y-%m'), last_12['Price'])
plt.title('Gold Price - Last 12 Months')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Price (USD per troy ounce)')
plt.tight_layout()
plt.show()

monthly_avg = df[df['Year'] >= 2015].groupby('Month')['Price'].mean()
plt.figure(figsize=(8,5))
plt.bar(monthly_avg.index, monthly_avg.values, color='goldenrod')
plt.title('Average Gold Price by Calendar Month (2015-Present)')
plt.xlabel('Month')
plt.ylabel('Average Price (USD)')
plt.show()

overall_growth = ((df['Price'].iloc[-1] - df['Price'].iloc[0]) / df['Price'].iloc[0]) * 100
print(f"Overall growth from {df['Date'].iloc[0].date()} to {df['Date'].iloc[-1].date()}: {overall_growth:.1f}%")

recent_5yr = df[df['Year'] >= df['Year'].max() - 5]
recent_growth = ((recent_5yr['Price'].iloc[-1] - recent_5yr['Price'].iloc[0]) / recent_5yr['Price'].iloc[0]) * 100
print(f"Growth over last 5 years: {recent_growth:.1f}%")
```
